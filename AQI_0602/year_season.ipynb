```python
import pandas as pd
avg = pd.read_csv('taipei_season_average.csv')#各季
avg['季節'].unique()
dict1 = {'冬季':1, '春雨':2, '梅雨':3, '夏季':4, '秋季':5}
avg['季節'].replace(dict1, inplace=True)
avg = avg[['年分','季節','平均氣溫[0C]','平均相對溼度[%]','日照時數[小時]','降水量[毫米]','降水日數[日]']]
avg
```

```python
avg.corr()
```

```python
import matplotlib.pyplot as plt 

season = avg['季節']
tem = avg['平均氣溫[0C]']
plt.plot(season, tem,color="red", lw="2.0", ls="--", label="label")
plt.legend()
plt.show()
```
