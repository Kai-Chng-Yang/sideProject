```python
#Z分數標準化
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('weather1.csv')
std = StandardScaler()
taipei = df[['溫度','風級','相對濕度']]
arr = std.fit_transform(taipei)
df[['溫度','風級','相對濕度']] = arr
df
```

```python
#皮爾森相關係數
import pandas as pd
df = pd.read_csv('weather1.csv')
taipei = df[['溫度','風級','相對濕度']]
featuresCorr = taipei.corr()
featuresCorr
```
